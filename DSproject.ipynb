**a) Exploratory Data Analysis**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('AAPL.csv')

# Convert Date to datetime and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Display basic info
print(df.info())
print(df.describe())
```

**1. Time Series Plot of Closing Price**

```python
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Closing Price')
plt.title('AAPL Stock Price (1980-1986)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

**Explanation:** This line plot shows the historical closing prices of AAPL stock. We can observe:

The stock started trading at around $0.13 in December 1980

Significant volatility in the early years

A major price increase in early 1983

A sharp decline in late 1983

Gradual recovery through 1984-1986



**2. Volume Analysis**

```python
plt.figure(figsize=(14, 7))
plt.plot(df['Volume'], label='Trading Volume', color='orange')
plt.title('AAPL Trading Volume (1980-1986)')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

**Explanation**: This shows the trading volume over time. Key observations:

Extremely high volume on the first trading day (IPO)

Periodic spikes in volume corresponding to major price movements

Generally higher volume during periods of price volatili

**3. Daily Price Range Analysis**

```python
df['Daily Range'] = df['High'] - df['Low']
plt.figure(figsize=(14, 7))
plt.plot(df['Daily Range'], label='Daily Price Range', color='green')
plt.title('AAPL Daily Price Range (1980-1986)')
plt.xlabel('Date')
plt.ylabel('Price Range ($)')
plt.legend()
plt.grid(True)
plt.show()
```

**Explanation**:This shows the difference between daily high and low prices. We can see:

The price range was very small in the early years (fractions of a cent)

Range expanded significantly during volatile periods

Largest ranges occurred during the 1983 price surge and subsequent correction

**4. Monthly Average Closing Price**

```python
monthly_avg = df['Close'].resample('M').mean()
plt.figure(figsize=(14, 7))
monthly_avg.plot(label='Monthly Average Close')
plt.title('AAPL Monthly Average Closing Price (1980-1986)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

**Explanation**: This smooths out the daily fluctuations to show monthly trends. Key points:

Steady rise from 1980 to early 1983

Sharp peak in January 1983

Significant correction through 1983

Gradual recovery through 1984-1986

**b) Histograms and Box Plots**

First 3 Attributes (Open, High, Low)

```python
# Histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df['Open'].hist(ax=axes[0], bins=50, color='blue')
axes[0].set_title('Open Price Distribution')
df['High'].hist(ax=axes[1], bins=50, color='green')
axes[1].set_title('High Price Distribution')
df['Low'].hist(ax=axes[2], bins=50, color='red')
axes[2].set_title('Low Price Distribution')
plt.tight_layout()
plt.show()

# Box plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df['Open'].plot(kind='box', ax=axes[0], vert=False)
axes[0].set_title('Open Price')
df['High'].plot(kind='box', ax=axes[1], vert=False)
axes[1].set_title('High Price')
df['Low'].plot(kind='box', ax=axes[2], vert=False)
axes[2].set_title('Low Price')
plt.tight_layout()
plt.show()
```

Last 3 Attributes (Adj Close, Volume, Daily Range)

```python
# Histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df['Adj Close'].hist(ax=axes[0], bins=50, color='purple')
axes[0].set_title('Adj Close Distribution')
df['Volume'].hist(ax=axes[1], bins=50, color='orange')
axes[1].set_title('Volume Distribution')
df['Daily Range'].hist(ax=axes[2], bins=50, color='cyan')
axes[2].set_title('Daily Range Distribution')
plt.tight_layout()
plt.show()

# Box plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df['Adj Close'].plot(kind='box', ax=axes[0], vert=False)
axes[0].set_title('Adj Close')
df['Volume'].plot(kind='box', ax=axes[1], vert=False)
axes[1].set_title('Volume')
df['Daily Range'].plot(kind='box', ax=axes[2], vert=False)
axes[2].set_title('Daily Range')
plt.tight_layout()
plt.show()
```

**Explanation**: The histograms show the distribution of values for each attribute, while the box plots show the median, quartiles, and outliers. Key observations:

Price attributes (Open, High, Low, Close, Adj Close) have similar distributions

Volume is extremely right-skewed with many outliers

Daily Range shows most days had small price movements with some extreme outliers

**c) Correlation Matrix Plot**

```python
# Calculate correlations
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of AAPL Stock Data')
plt.show()
```

**Explanation**: The correlation matrix shows:

Price attributes (Open, High, Low, Close, Adj Close) are nearly perfectly correlated (≈1.0)

Volume has weak negative correlation with prices (-0.1 to -0.2)

Daily Range has moderate positive correlation with prices (0.4-0.5)

**d) Density Plots of Last 3 Attributes**

```python
# Density plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df['Adj Close'].plot(kind='density', ax=axes[0])
axes[0].set_title('Adj Close Density')
df['Volume'].plot(kind='density', ax=axes[1])
axes[1].set_title('Volume Density')
df['Daily Range'].plot(kind='density', ax=axes[2])
axes[2].set_title('Daily Range Density')
plt.tight_layout()
plt.show()
```

**Explanation**: The density plots show the probability distribution of values:

Adj Close has a multi-modal distribution reflecting different price regimes

Volume is extremely right-skewed, with most days having lower volume

Daily Range shows most days had very small price movements with a long right tail
